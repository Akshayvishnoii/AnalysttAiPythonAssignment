```python
import requests
import string
import re
import pandas as pd
import urllib
```

```python
url = "https://www.amazon.in/s?k=bags&crid=2M096C61O4MLT&qid=1653308124&sprefix=ba%2Caps%2C283&ref=sr_pg_1"
```

```python
response = requests.get(url, headers={"user-Agent": "XY"})
```

```python
response.status_code
```

```python
page_contents = response.text
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(page_contents, 'html.parser')
```

```python
print(soup.prettify())
```

```python
for link in soup.find_all('a', attrs={'href': re.compile("^https")}):
    print(link)
type(link)
```

```python
product_urls = []
product_names = []
product_prices = []
ratings = []
num_reviews = []
```

```python
for page in range(1, 21):
 results = soup.find_all('div', {'data-component-type': 's-search-result'})
 for result in results:
        product_url = result.find('a', {'class': 'a-link-normal s-no-outline'})['href']
        product_name = result.find('span', {'class': 'a-text-normal'}).text
        product_price = result.find('span', {'class': 'a-price-whole'}).text
        rating = result.find('span', {'class': 'a-icon-alt'}).text.split()[0]
        num_review = result.find('span', {'class': 'a-size-base'}).text
        product_urls.append(product_url)
        product_names.append(product_name)
        product_prices.append(product_price)
        ratings.append(rating)
        num_reviews.append(num_review)
```

```python
data = {'Product URL': product_urls, 'Product Name': product_names, 'Product Price': product_prices, 'Rating': ratings, 'Number of Reviews': num_reviews}
df = pd.DataFrame(data)
```

```python
import csv
```

```python
product_data = []
```

```python
 # extract the additional information from the product page
 asin = soup.find('th', {'class': 'a-color-secondary a-size-base prodDetSectionEntry'})
 if asin is not None:
  asin = asin.text.strip()

 product_description = soup.find('div', {'id': 'productDescription'})
 if product_description is not None:
  product_description = product_description.text.strip()

 manufacturer = soup.find('a', {'id': 'bylineInfo'})
 if manufacturer is not None:
   manufacturer = manufacturer.text.strip()

 # add the product data to the list
 product_data.append([product_url, product_name, product_price, rating, num_reviews, asin, product_description, manufacturer])

# print out the extracted data
print(product_url, product_name, product_price, rating, num_reviews, asin, product_description, manufacturer)

# write the data to a CSV file
with open('products.csv', 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Product URL', 'Product Name', 'Product Price', 'Rating', 'Number of Reviews', 'ASIN', 'Product Description', 'Manufacturer'])
    writer.writerows(product_data)
```
